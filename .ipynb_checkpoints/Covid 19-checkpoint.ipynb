```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings 
import seaborn as sns
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('country_wise_latest.csv')
```

```python
df.head(3)
```

```python
df.rename(columns={'Country/Region':'Country' } , inplace=True)
```

```python
df.head(2)
```

```python
df.set_index('Country',inplace=True)
```

```python
top = df['Deaths / 100 Recovered'].nlargest(10)
```

```python
sort = df.sort_values(by='Deaths / 100 Cases',ascending=False)
sort_top = sort.head(10)
```

```python
sort_top.set_index('Country',inplace=True)
```

```python
plt.figure(figsize=(15,7))
sort_top['Deaths / 100 Cases'].plot(kind='pie',autopct='%.1f%%')
plt.title('Death Rate In Top 10 Country')
```

```python
plt.figure(figsize=(15,9))
sort1 = df.sort_values(by='Recovered / 100 Cases',ascending=False)
sort_top1 = sort1.head(10)
sort_top1.set_index('Country',inplace=True)
plt.figure(figsize=(15,7))
sort_top['Recovered / 100 Cases'].plot(kind='pie',autopct='%.1f%%')
plt.title('Recover Rate In Top 10 Country')
plt.ylabel(' ')
plt.show()
```

```python
plt.figure(figsize=(15,7))
df['WHO Region'].value_counts().sort_values().plot(kind='pie' , autopct='%.1f%%')
plt.ylabel(' ')
plt.title('Covid 19 in Each Region')
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(15,7))
df['WHO Region'].value_counts().sort_values().plot(kind='bar')
plt.xticks(rotation=50)
plt.ylabel(' ')
plt.title('Covid 19 in Each Region')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.set_style('darkgrid')
top2 = df.sort_values(by='Confirmed',ascending=False)
top2 = top2.head(10)
top2
sns.barplot(x=top2['Confirmed'] , y=top2['Country'],palette='BrBG')
plt.title('Top 10 Country Confirmed Cases')
plt.yticks(rotation=10)
plt.show()
```
